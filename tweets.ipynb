```python

def exactPhrase(sval, uval, keyword='', maxTweets = 100):
    twtt = []
    for i, tweet in enumerate(
            twitter.TwitterSearchScraper(
                keyword + 
                ' since:' +sval + '' + 'until:' +uval + 'lang:"en" ').get_items()):
        if i > maxTweets:
            break
        tweets = {
            "Tweet" : "Reply",
            "tweet/reply id": tweet.id,
            "inReplyToTweetId": tweet.inReplyToTweetId,
            "conversationId": tweet.conversationId,
            "tweet.username" : tweet.user.username,
            "tweet.content" : tweet.rawContent,
            "tweet.date" : tweet.date,
            "tweet.user.location" : tweet.user.location,
            "tweet.likeCount" : tweet.likeCount, 
            "tweet.replyCount" : tweet.replyCount,
            "tweet.retweetCount" : tweet.retweetCount,
            "tweet.user.followersCount" : tweet.user.followersCount,
            "tweet.user.description": tweet.user.renderedDescription,
            "tweet.user.friendsCount": tweet.user.friendsCount,
            "tweet.user.statusesCount": tweet.user.statusesCount,
            "tweet.user.favouritesCount": tweet.user.favouritesCount,
            "tweet.user.listedCount": tweet.user.listedCount,
            "tweet.user.mediaCount": tweet.user.mediaCount,
            "tweet.url" : tweet.url
            }
        twtt.append(tweets)
    return twtt

```

```python
nft = exactPhrase('2022-12-01', '2022-12-08' '"nft"', maxTweets=7000)
```

```python
# Creating list to append tweet data 
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(twitter.TwitterSearchScraper('from:inecnigeria since:2020-01-01 until:2023-03-01').get_items()): #declare a username 
    if i>100000: #number of tweets you want to scrape
        break
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username]) #declare the attributes to be returned
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
twt_df = tweets_df1.copy()
twt_df.head()
```

```python
def remove_punctuation(text):
    final = "".join(u for u in text if u not in ("\n"))
    return final
```

```python
twt_df['Text'] = twt_df['Text'].apply(remove_punctuation)

twt_df.head()
```
